```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv("/content/Q1_clean.csv")
df.columns = df.columns.str.strip()
df["Log Price"] = np.log(df["Price"])


# 1) Average Prices and Scores by Neighbourhood

avg_by_neighbourhood = df.groupby("Neighbourhood")[["Price", "Review Scores Rating"]].mean().sort_values(by="Price", ascending=False)
print("Average Price and Review Score by Neighbourhood:\n", avg_by_neighbourhood)
most_expensive = avg_by_neighbourhood.index[0]
print(f"\nMost expensive borough: {most_expensive}")

# 2) Kernel Density Plots
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.kdeplot(data=df, x="Price", hue="Neighbourhood", common_norm=False)
plt.title("Kernel Density Plot of Price by Neighbourhood")
plt.xlim(0, 1000)
plt.show()
plt.figure(figsize=(12, 6))
sns.kdeplot(data=df, x="Log Price", hue="Neighbourhood", common_norm=False)
plt.title("Kernel Density Plot of Log(Price) by Neighbourhood")
plt.show()


# 3) Regression: Price, Neighbourhood (No Intercept)
X_no_intercept = pd.get_dummies(df['Neighbourhood'], drop_first=False)
y = df['Price']
model_no_intercept = sm.OLS(y, X_no_intercept).fit()
print("\n3. Regression without Intercept:\n", model_no_intercept.summary())


# 4) Regression: Price, Neighbourhood
X_dummies_intercept = pd.get_dummies(df['Neighbourhood'], drop_first=True).astype(float)
X_with_intercept = sm.add_constant(X_dummies_intercept)
y = df['Price'].astype(float)

model_with_intercept = sm.OLS(y, X_with_intercept).fit()
print("\n4. Regression with Intercept:\n", model_with_intercept.summary())

# 5) Regression: Price, Review Score, Neighbourhood
X_part5 = pd.get_dummies(df[['Review Scores Rating', 'Neighbourhood']], drop_first=True)
y = df['Price']
X_train_5, X_test_5, y_train_5, y_test_5 = train_test_split(X_part5, y, test_size=0.2, random_state=42)

model_part5 = LinearRegression()
model_part5.fit(X_train_5, y_train_5)

y_pred_5 = model_part5.predict(X_test_5)
rmse_5 = np.sqrt(mean_squared_error(y_test_5, y_pred_5))
r2_5 = r2_score(y_test_5, y_pred_5)
coef_rating_5 = model_part5.coef_[X_part5.columns.get_loc('Review Scores Rating')]

print("\n5. Regression with Review Scores + Neighbourhood")
print(f"RMSE: {rmse_5:.2f}, R²: {r2_5:.4f}, Coefficient on Review Score: {coef_rating_5:.4f}")


# 6) Regression: Add Property Type
X_part6 = pd.get_dummies(df[['Review Scores Rating', 'Neighbourhood', 'Property Type']], drop_first=True)
X_train_6, X_test_6, y_train_6, y_test_6 = train_test_split(X_part6, y, test_size=0.2, random_state=42)

model_part6 = LinearRegression()
model_part6.fit(X_train_6, y_train_6)

y_pred_6 = model_part6.predict(X_test_6)
rmse_6 = np.sqrt(mean_squared_error(y_test_6, y_pred_6))
r2_6 = r2_score(y_test_6, y_pred_6)
coef_rating_6 = model_part6.coef_[X_part6.columns.get_loc('Review Scores Rating')]

# most expensive property type
property_type_coefs = pd.Series(model_part6.coef_, index=X_part6.columns)
base_price = model_part6.intercept_
property_type_columns = [col for col in X_part6.columns if col.startswith("Property Type")]
most_expensive_property = property_type_coefs[property_type_columns].idxmax()
most_expensive_price = base_price + property_type_coefs[most_expensive_property]

print("\n6. Regression with Property Type")
print(f"RMSE: {rmse_6:.2f}, R²: {r2_6:.4f}, Coefficient on Review Score: {coef_rating_6:.4f}")
print(f"Most expensive property type: {most_expensive_property} → Est. Price: ${most_expensive_price:.2f}")
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

heart_df = pd.read_csv("/content/heart_hw.csv")

# Convert transplant column to binary: 0 = control, 1 = treatment
heart_df['transplant_binary'] = heart_df['transplant'].map({'control': 0, 'treatment': 1})

# 1) Proportion Calculations
# (a) Proportion surviving in control group
control_survival_rate = heart_df[heart_df['transplant_binary'] == 0]['y'].mean()

# (b) Proportion surviving in treatment group and ATE
treatment_survival_rate = heart_df[heart_df['transplant_binary'] == 1]['y'].mean()
ate = treatment_survival_rate - control_survival_rate

print("1. Proportions")
print(f"Control group survival rate: {control_survival_rate:.4f}")
print(f"Treatment group survival rate: {treatment_survival_rate:.4f}")
print(f"Average Treatment Effect (ATE): {ate:.4f}\n")

# 2) Linear Regression: y, transplant
X_simple = sm.add_constant(heart_df['transplant_binary'])
y = heart_df['y']
model_simple = sm.OLS(y, X_simple).fit()
print("2. Regression: y ~ transplant")
print(model_simple.summary(), "\n")

# 3) Linear Regression: y, transplant, age
X_with_age = sm.add_constant(heart_df[['transplant_binary', 'age']])
model_with_age = sm.OLS(y, X_with_age).fit()
print("3. Regression: y ~ transplant + age")
print(model_with_age.summary(), "\n")

# 4) Flexible Model: Add age and interaction
heart_df['age_squared'] = heart_df['age'] ** 2
heart_df['age_transplant'] = heart_df['age'] * heart_df['transplant_binary']

# features and target
features = ['transplant_binary', 'age', 'age_squared', 'age_transplant']
X = heart_df[features]
y = heart_df['y']

# train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# logistic regression
model_flex = LogisticRegression()
model_flex.fit(X_train, y_train)

# Predict over a range of ages for both groups
ages = np.linspace(heart_df['age'].min(), heart_df['age'].max(), 100)
df_plot = pd.DataFrame({
    'age': np.tile(ages, 2),
    'transplant_binary': np.repeat([0, 1], len(ages))
})
df_plot['age_squared'] = df_plot['age'] ** 2
df_plot['age_transplant'] = df_plot['age'] * df_plot['transplant_binary']

# Predict probabilities
probs = model_flex.predict_proba(df_plot[features])[:, 1]
df_plot['predicted_survival'] = probs

plt.figure(figsize=(10, 6))
for label, group in df_plot.groupby('transplant_binary'):
    label_name = "Treatment" if label == 1 else "Control"
    plt.plot(group['age'], group['predicted_survival'], label=label_name)

plt.xlabel("Age")
plt.ylabel("Predicted Probability of Survival")
plt.title("Predicted Survival Probability by Age and Treatment Group")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Q10.** Let's look at a cousin of Linear Regression, called **kernel regression** or **local constant least squares** or **Nadaraya-Watson Estimator**.

We derived the OLS estimator for single linear regression by minimizing
$$
SSE(b_0, b_1) = \frac{1}{N}\sum_{i=1}^N (y_i - b_0 - b_1 x_i)^2
$$
with solution
$$
\hat{b}_0 = \bar{y} - \hat{b}_1 \bar{x}, \quad \hat{b}_1 = \dfrac{\frac{1}{N} \sum_{i=1}^N (x_i-\bar{x})(y_i - \bar{y})}{s_x^2}.
$$

When you step back and think about it, this is a bit weird: The algorithm is computing sample means, variances, and covariances, and using those to create a predictive model. The data themselves arguably vanish from the solution. This is elegant, this is strange.

Instead, let $k(z)$ be a kernel function, such as the Gaussian
$$
k(z) = \frac{1}{\sqrt{2\pi}} e^{-z^2/2}
$$
or uniform
$$
k(z) = \begin{cases}
1/2, & |z| \le \frac{1}{2}\\
0, & \text{otherwise.}
\end{cases}
$$
We'll instead minimize, or each predictor value $x$,
$$
SSE(\hat{y}(x)) = \dfrac{1}{N} \sum_{i=1}^N \left\lbrace y_i - \hat{y}(x)\right\rbrace^2 \dfrac{1}{h} k \left( \dfrac{x-x_i}{h} \right).
$$

1. Show that the optimal predictor is
$$
\hat{y}(x) = \dfrac{ \frac{1}{Nh} \sum_{i=1}^N  y_i k \left( \dfrac{x-x_i}{h} \right) }{\frac{1}{Nh} \sum_{i=1}^N k \left(  \dfrac{x-x_i}{h} \right)}
$$
This has many names, but let's call it the local constant least squares (LCLS) estimator, as opposed to ordinary least squares (OLS).
